```python
url = "https://www.alltrails.com/parks/us/colorado/rocky-mountain-national-park"
response = urllib.request.urlopen(url)
html = response.read()
alltrails = BeautifulSoup(html, "lxml")
alltrails
```

```python
comments = alltrails.findAll("p", {"itemprop": "reviewBody"})
comments_clean = []
for i in comments:
    comments_clean.append(i.getText())

comments_clean
```

```python
comments_tokenized = []
for i in comments_clean:
    tknzr = TweetTokenizer(strip_handles=True, preserve_case=False)
    tokens = tknzr.tokenize(i)
    tokens = [w for w in tokens if w not in string.punctuation]
    tokens = [w for w in tokens if w not in nltk.corpus.stopwords.words('english')]
    tokens = [w for w in tokens if len(w) > 2]
    tokens = [w for w in tokens if w[0:4] != "http"]
    tokens = [w for w in tokens if w[0] != "#"]
    comments_tokenized.append(tokens)

comments_tokenized
```

```python
ratings = alltrails.findAll("meta", {"itemprop": "ratingValue"}, {"data-reactid": "77"})
ratings = ratings[1:31]
ratings_clean = []
for i in ratings:
    ratings_clean.append(i.attrs["content"])

ratings_clean
```

```python
alltrails_comments = list(zip(comments_tokenized,ratings_clean))
alltrails_comments
```

# Model

```python
words = []
for i in comments_tokenized:
    words.append((" ").join(i))

word_features = []
word_features = (" ").join(words)
word_features = word_features.split(" ")
word_features = list(set(word_features))
word_features
```

```python
def extract_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains(%s)' % word] = (word in document_words)
    return features
```

```python
training_set = nltk.classify.util.apply_features(extract_features, alltrails_comments)
print (training_set)
```

```python
testing_set = nltk.classify.util.apply_features(extract_features, tweets_tagged)
print (testing_set)
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
print('accuracy:', nltk.classify.util.accuracy(classifier, testing_set))
classifier.show_most_informative_features()
```
